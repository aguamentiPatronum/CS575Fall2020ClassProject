```python
import pandas as pd
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# To use, just change the from and to variables. 
# Will then process the DFs to produce separate parquet files
# That only have the actions from the evaluation

from_file_loc = '/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/PowerPillsOnlyEvalDB.parquet'
actionCounts_file_loc = '/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/PowerPills_Actions.parquet'
actionPercents_file_loc = '/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/PowerPills_ActionPercentages.parquet'
```

```python
df = pd.read_parquet(from_file_loc)
df = df.drop(columns=df.columns[0])
```

```python
df.columns
```

```python
# Can I get the sums of each action for each epoch?
actions_df = pd.DataFrame(df.pivot_table(index='epoch',columns='action_name',aggfunc=sum))
df.head(3)
```

```python
action_subset = pd.DataFrame(actions_df['action'])
action_subset['epoch'] = range(1, len(action_subset) + 1)
action_subset.set_index('epoch',drop=False,inplace=True)
action_subset
```

```python
actions_only = action_subset.drop(['epoch'], axis=1)

plt.stackplot(range(1,len(actions_only)+1), actions_only['DOWN'], actions_only['LEFT'], actions_only['RIGHT'], actions_only['UP'], labels = ['DOWN', 'LEFT', 'RIGHT', 'UP'])

plt.title('Actions Per Life')
plt.xlabel('Life')
plt.ylabel('Number of Actions Taken in Life')
plt.legend(loc="lower left")
```

```python
data_perc = actions_only.divide(actions_only.sum(axis=1), axis=0)
data_perc
plt.stackplot(range(1,len(data_perc)+1), data_perc['DOWN'], 
              data_perc['LEFT'], 
              data_perc['RIGHT'], 
              data_perc['UP'], 
              labels = ['DOWN', 'LEFT', 'RIGHT', 'UP'])
plt.title('Actions Per Life')
plt.xlabel('Life')
plt.ylabel('Percent of Actions Taken in Life')
plt.legend(loc="lower left")
```

```python

table = pa.Table.from_pandas(actions_only)
# Parquet with Brotli compression
pq.write_table(table, actionCounts_file_loc, compression='BROTLI')

table = pa.Table.from_pandas(data_perc)
# Parquet with Brotli compression
pq.write_table(table, actionPercents_file_loc, compression='BROTLI')
```
