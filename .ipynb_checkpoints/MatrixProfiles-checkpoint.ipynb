```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matrixprofile as mp
%matplotlib inline
import stumpy as sp
```

```python
FearDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/FearGhostsTrainingDB.parquet')
FearDF = FearDF.drop(columns=FearDF.columns[0])
NormalOnlyDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/NormalPillsOnlyDB.parquet')
NormalOnlyDF = NormalOnlyDF.drop(columns=NormalOnlyDF.columns[0])
NormalPlusDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/NormalPillsInGameDB.parquet')
NormalPlusDF = NormalPlusDF.drop(columns=NormalPlusDF.columns[0])
PowerDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/PowerPillsOnlyDB.parquet')
PowerDF = PowerDF.drop(columns=PowerDF.columns[0])
StdDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/StandardScoringOnlyDB.parquet')
StdDF = StdDF.drop(columns=StdDF.columns[0])

train_df_array = [FearDF, NormalOnlyDF, NormalPlusDF, PowerDF, StdDF]
names_array = ['fear', 'normalOnly', 'normalPlus', 'power', 'standard']
```

```python
FearEvalDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/FearGhostsEvalDB.parquet')
FearEvalDF = FearEvalDF.drop(columns=FearEvalDF.columns[0])
NormalOnlyEvalDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/NormalPillsOnlyEvalDB.parquet')
NormalOnlyEvalDF = NormalOnlyEvalDF.drop(columns=NormalOnlyEvalDF.columns[0])
# NormalPlusEvalDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/TrainingSlimmed/NormalPillsInGameDB.parquet')
# NormalPlusEvalDF = NormalPlusEvalDF.drop(columns=NormalPlusEvalDF.columns[0])
PowerEvalDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/PowerPillsOnlyEvalDB.parquet')
PowerEvalDF = PowerEvalDF.drop(columns=PowerEvalDF.columns[0])
StdEvalDF = pd.read_parquet('/Volumes/Britt_SSD/Singularity/EvaluationsSlimmed/StandardEval2DB.parquet')
StdEvalDF = StdEvalDF.drop(columns=StdEvalDF.columns[0])

eval_df_array = [FearEvalDF, NormalOnlyEvalDF, PowerEvalDF, StdEvalDF]
names_array = ['fear', 'normalOnly', 'power', 'standard']

```

```python
fearEval_actions_sequence = np.array(FearEvalDF['action'])
```

```python
fearEval_actions_profile = mp.analyze(fearEval_actions_sequence)
```

```python
fearEval_actions_profile = mp.compute(fearEval_actions_sequence, windows=10)
fearEval_actions_profile = mp.discover.motifs(fearEval_actions_profile, k=1)
figures = mp.visualize(fearEval_actions_profile)
```

```python
# Can we use stumpy to compare two evals?
m1 = 18
queen_mp = sp.stump(T_A = FearEvalDF['action'].astype(float),
                  m = m1,
                  T_B = NormalOnlyEvalDF['action'].astype(float),
                  ignore_trivial = False)
```

```python
queen_motif_index = queen_mp[:, 0].argmin()
```

```python

plt.xlabel('Subsequence')
plt.ylabel('Matrix Profile')

plt.scatter(queen_motif_index,
               queen_mp[queen_motif_index, 0],
               c='red',
               s=100)

plt.plot(queen_mp[:,0])

plt.show()
```

```python
# Now, to discover the global motif (i.e., the most conserved pattern), queen_motif_index, all we need to do is identify the index location of the lowest distance value in the queen_mp matrix profile (see red circle above).
queen_motif_index = queen_mp[:, 0].argmin()
print(f'The motif is located at index {queen_motif_index} of "FearEvalDF"')
```

```python
#In fact, the index location of its nearest neighbor in Ice Ice Baby is stored in queen_mp[queen_motif_index, 1]:
vanilla_ice_motif_index = queen_mp[queen_motif_index, 1]
print(f'The motif is located at index {vanilla_ice_motif_index} of "NormalOnlyEvalDF"')
```

```python
#Overlaying The Best Matching Motif
plt.plot(fearEval_actions_sequence[queen_motif_index:queen_motif_index + m1],label='Fear Eval',linestyle="--")
plt.plot(normalOnlyEval_actions_sequence[vanilla_ice_motif_index:vanilla_ice_motif_index+m1],label='Normal Only Eval',linestyle=":",alpha=0.8)

plt.xlabel('Time')
plt.ylabel('Action Sequence')

plt.legend()

plt.show()
```
