```python
import pandas as pd
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
complete_df = pd.DataFrame()
tempTopDir = os.path.abspath('/Volumes/Britt_SSD/Singularity/Location_Evals/FearGhosts')
```

```python
agent_name = "fearGhost"
```

```python
# This notebook adds several columns to the EVAL dbs
# Just change the input file location and should run
from_file = os.path.join(tempTopDir, 'df1main.parquet')
to_parquet_file = os.path.join(tempTopDir, 'fearGhost_temp.parquet')
to_csv_file = os.path.join(tempTopDir, 'fearGhost_temp.csv')
df = pd.read_parquet(from_file)
df = df.drop(columns=df.columns[0])
```

```python
df.columns
```

```python
scores_path = os.path.join(tempTopDir, 'scores.txt')
        
f = open(scores_path, 'r')
data_txt = f.read()
data_txt = data_txt.replace('[', '')
data_txt = data_txt.replace(']', '')
epoch_scores = np.fromstring(data_txt, dtype=float, sep=',')
print("scores")
for item in epoch_scores:
    print(item)
```

```python
import_temp = pd.read_csv(os.path.join(tempTopDir, 'state_features_impoartance.csv'), index_col=0)
print(import_temp.columns)
print(complete_df.columns)
import_temp.sort_values('state',inplace=True)

complete_df = pd.merge(left=df, right=import_temp, on='state')
print(complete_df.columns)
```

```python
indices = complete_df.index[complete_df['end_of_epoch']].tolist()
indices.append(len(complete_df) - 1)
print("indices")
print(indices)
```

```python
key_states = np.load(os.path.join(tempTopDir, 'summary_states.npy')).tolist()
temp_context_states = np.load(os.path.join(tempTopDir, 'summary_states_with_context.npy'))
print(key_states)
print(temp_context_states)
context_states = [i for i in temp_context_states if i not in key_states]
```

```python
complete_df['epoch_score'] = 0
complete_df['key_state'] = False
complete_df['context_state'] = False

#if we are going to iterate, let's just iterate once
all_indices = np.hstack((indices, key_states, context_states))
print("all indices; ")
print(all_indices)
```

```python
for state in all_indices:
    if ((state < len(complete_df)) and (state > 0)) :
        # if it is the end of an epoch, fill in the score
        if state in indices:
            #print("Found item in indices")
            list_index = indices.index(state)
            if (complete_df.iloc[state]['epoch_score'] == 0):
                complete_df['epoch_score'][state] = epoch_scores[list_index-1]
                #print(complete_df.iloc[state]['epoch_score'])
        # if is a context state, fill in that column
        if state in context_states:
#             print("found item in context and state it: " + str(state))
#             print("columns are ")
#             print(complete_df.columns)
#             print("DF length: ")
#             print(len(complete_df))
#             print("Context States are: ")
#             print(context_states)
            context_index = context_states.index(state)
            print(complete_df.iloc[state])
            if (complete_df.iloc[state]['context_state'] == False):
#                 print("false found")
                complete_df['context_state'][state] = True
#                 print(complete_df.iloc[state]['context_state'])
        # if is a key state, fill in that column
        if state in key_states:
#             print("found item in keys")
            key_index = key_states.index(state)
            if (complete_df.iloc[state]['key_state'] == False):
#                 print("false found")
                complete_df['key_state'][state] = True
#                 print(complete_df.iloc[state]['key_state'])
                
```

```python
complete_df['action 0 episode sum']
```

```python
complete_df.drop(["action 0 episode sum", "action 0 total sum", "action 5 episode sum", "action 5 total sum", "action 6 episode sum", "action 6 total sum", "action 7 episode sum", "action 7 total sum", "action 8 episode sum", "action 8 total sum"], axis=1, inplace=True)
```

```python
# We want to add a column to the dataframe which numbers the key state
# that each context or key state belongs to, so first we make a new
#df with only the key and context states
key_statesDF = complete_df[complete_df.key_state.eq(True) | complete_df.context_state.eq(True)]
key_statesDF.tail(3)
```

```python
# Add new column which says which key state each item belongs to
key_statesDF['tempKeyNum'] = key_statesDF.groupby([key_statesDF.state, key_statesDF.state.diff().ne(1).cumsum()])
key_statesDF['tempKeyNum2'], temp2 = zip(*key_statesDF['tempKeyNum'])
temp2, key_statesDF['keyNum'] = zip(*key_statesDF['tempKeyNum2'])
key_statesDF = key_statesDF.drop(columns=['tempKeyNum','tempKeyNum2'])
print(key_statesDF['keyNum'])
```

```python
# Now to add that information back to the original DB
# If the column has a 0, step is not key or context state
df = pd.concat([complete_df, key_statesDF['keyNum']], axis=1)
```

```python
# Replace NaN with 0
df['keyNum'] = df['keyNum'].fillna(0)
df[df['keyNum'] == 1]
```

```python
df.drop(["q_values", "features"], axis=1, inplace=True)
df.columns
```

```python
# Now we will mark the five steps just before losing a life
idx = df.index.get_indexer_for(df[df.end_of_episode==True].index)
n=5
#key_statesNormal.iloc[np.unique(np.concatenate([np.arange(max(i-n,0), min(i+n+1, len(key_statesNormal)))
                                         #   for i in idx]))]
fiveBeforeLifeLoss = df.iloc[np.unique(np.concatenate([np.arange(max(i-n,0), min(i+n-4, len(df)))
                                            for i in idx]))]

fiveBeforeLifeLoss['beforeLifeLoss'] = True
print(fiveBeforeLifeLoss.head(3))
```

```python
# Now to add that information back to the original DB
# If the column has a 0, step is not in the group of five before each life loss
df = pd.concat([df, fiveBeforeLifeLoss['beforeLifeLoss']], axis=1)
# Replace NaN with 0
df['beforeLifeLoss'] = df['beforeLifeLoss'].fillna(False)
df.head(1)
```

```python
# And the five steps leading up to any reward larger than 10
idx = df.index.get_indexer_for(df[df.reward>10].index)
n=5
fiveAroundBigRewards = df.iloc[np.unique(np.concatenate([np.arange(max(i-n,0), min(i+n+1, len(df)))
                                            for i in idx]))]
fiveAroundBigRewards['aroundBigReward'] = True
fiveAroundBigRewards.head(1)
```

```python
fiveAroundBigRewards['tempKeyNum'] = fiveAroundBigRewards.groupby([fiveAroundBigRewards.state, fiveAroundBigRewards.state.diff().ne(1).cumsum()])
fiveAroundBigRewards['tempKeyNum2'], temp2 = zip(*fiveAroundBigRewards['tempKeyNum'])
temp2, fiveAroundBigRewards['bigRewardNum'] = zip(*fiveAroundBigRewards['tempKeyNum2'])
fiveAroundBigRewards = fiveAroundBigRewards.drop(columns=['tempKeyNum','tempKeyNum2'])
print(fiveAroundBigRewards['bigRewardNum'])
```

```python
# Now to add that information back to the original DB
# If the column has a 0, step is not key or context state
df = pd.concat([df, fiveAroundBigRewards['bigRewardNum']], axis=1)
# Replace NaN with 0
df['bigRewardNum'] = df['bigRewardNum'].fillna(0)
df.head(1)
```

```python
# Add a column which specifies which agent this is

df['agent'] = agent_name
df[df['bigRewardNum'] == 1]
```

```python
# Now add some averaging columns
# average distance to all four pills
col = df.loc[: , "to_pill_one":"to_pill_four"]
df['to_pill_mean'] = col.mean(axis=1)
```

```python
# average distance to top pills
col = df.loc[: , "to_pill_one":"to_pill_two"]
df['to_top_pills_mean'] = col.mean(axis=1)
```

```python
# and to bottom pills
col = df.loc[: , "to_pill_three":"to_pill_four"]
df['to_bottom_pills_mean'] = col.mean(axis=1)
```

```python
# average distance to all ghosts
col = df.loc[: , "to_red_ghost":"to_orange_gohst"]
df['to_ghosts_mean'] = col.mean(axis=1)
```

```python
# Distance to Blue Ghosts

subset = df[['pacman_coord_x', 'pacman_coord_y']]
pacman_coord = [np.array(x) for x in subset.to_numpy()]

subset = df[['dark_blue_ghost1_coord_x', 'dark_blue_ghost1_coord_y']]
db1_coord = [np.array(x) for x in subset.to_numpy()]

subset = df[['dark_blue_ghost2_coord_x', 'dark_blue_ghost2_coord_y']]
db2_coord = [np.array(x) for x in subset.to_numpy()]

subset = df[['dark_blue_ghost3_coord_x', 'dark_blue_ghost3_coord_y']]
db3_coord = [np.array(x) for x in subset.to_numpy()]

subset = df[['dark_blue_ghost4_coord_x', 'dark_blue_ghost4_coord_y']]
db4_coord = [np.array(x) for x in subset.to_numpy()]

to_db1 = []
to_db2 = []
to_db3 = []
to_db4 = []
    
for i in range(0,len(pacman_coord)):
    to_db1.append(np.abs(pacman_coord[i] - db1_coord[i]).sum())
    to_db2.append(np.abs(pacman_coord[i] - db2_coord[i]).sum())
    to_db3.append(np.abs(pacman_coord[i] - db3_coord[i]).sum())
    to_db4.append(np.abs(pacman_coord[i] - db4_coord[i]).sum())
                             
df['to_db1'] = to_db1
df['to_db2'] = to_db2
df['to_db3'] = to_db3
df['to_db4'] = to_db4
```

```python
table = pa.Table.from_pandas(df)
# Parquet with Brotli compression
pq.write_table(table, to_parquet_file, compression='BROTLI')
# And save to CSV for permanent storage
df.to_csv(to_csv_file)
```
